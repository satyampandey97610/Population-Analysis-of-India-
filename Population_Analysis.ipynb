
# Population Analysis of India

This notebook provides a comprehensive analysis of the population statistics of Indian states and union territories. The purpose of this analysis is to explore patterns and relationships in the data, presented in a clear and structured manner.

---
## Tools Used
- **Pandas**: For data manipulation and analysis.
- **NumPy**: For numerical computations.
- **Matplotlib**: For data visualization.
---



## Step 1: Loading the Dataset

Let's begin by loading the dataset and taking an initial look at its structure to understand the data better.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

file_path = "Population of India.csv"
data = pd.read_csv(file_path)

print("Dataset Info:")
data.info()
print("First Few Rows of the Dataset:")
print(data.head())
```


## Step 2: Descriptive Statistics

We will compute basic statistics to summarize the dataset, such as mean, median, and standard deviation. This helps in understanding the overall distribution of the data.


```python

print("Descriptive Statistics:")
print(data.describe())
```


## Step 3: Visualizations

Visualizations make it easier to understand and communicate patterns in the data. Here, we create simple plots to explore key aspects.


### Top 10 States by Population

```python

plt.figure(figsize=(10, 6))
top_population = data.sort_values(by="Population[50]", ascending=False).head(10)
plt.barh(top_population["State/UT"], top_population["Population[50]"], color="skyblue")
plt.title("Top 10 States by Population")
plt.xlabel("Population")
plt.ylabel("State/UT")
plt.show()
```

### Rural vs Urban Population Distribution

```python

plt.figure(figsize=(8, 8))
rural_urban_totals = [data["Rural[51]"].sum(), data["Urban[51]"].sum()]
labels = ["Rural", "Urban"]
plt.title("Rural vs Urban Population Distribution")
plt.show()
```

### Scatter Plot: Population Density vs Area

```python

plt.figure(figsize=(10, 6))
plt.scatter(data["Area[52] (km2)"], data["Density (per km2)"], color="purple", alpha=0.6)
plt.title("Population Density vs Area")
plt.xlabel("Area (km²)")
plt.grid(True)
plt.show()
```


## Step 4: Key Insights

Finally, let's derive some meaningful insights from the data, such as the states with the highest and lowest populations and an exploration of relationships between variables.


```python

highest_population = data.loc[data["Population[50]"].idxmax()]
lowest_population = data.loc[data["Population[50]"].idxmin()]
print("State with Highest Population:", highest_population["State/UT"], highest_population["Population[50]"])
print("State with Lowest Population:", lowest_population["State/UT"], lowest_population["Population[50]"])
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data["Rural[51]"], data["Urban[51]"], color="green", alpha=0.6)
plt.title("Rural vs Urban Population")
plt.xlabel("Rural Population")
plt.ylabel("Urban Population")
plt.grid(True)
plt.show()
```
